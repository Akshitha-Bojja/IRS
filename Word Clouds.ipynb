```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
synonyms = []
antonyms = []
for syn in wordnet.synsets("happy"):
            for l in syn.lemmas():
                   synonyms.append(l.name())
                   if l.antonyms():
                           antonyms.append(l.antonyms()[0].name())
print('Synonyms are :')
print(set(synonyms))
print('Antonyms are:')
print(set(antonyms))
```
